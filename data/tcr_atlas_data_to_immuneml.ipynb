```python
import scanpy as sc
import pandas as pd
import numpy as np
```

```python
tcr_adat = sc.read_h5ad("human_tcr_reference_v2.h5ad")
```

```python
df = pd.read_excel("SampleInformation_v2_5.xlsx", index_col=0)
df
```

```python
#standard AIRR tsv fieds
col_names_airr = ['sequence_id', 'sequence', 'quality', 'sequence_aa', 'rev_comp', 'productive', 'vj_in_frame', 'stop_codon', 'complete_vdj', 'locus', 'v_call', 'd_call', 'd2_call', 'j_call', 'c_call', 'sequence_alignment', 'quality_alignment', 'sequence_alignment_aa', 'germline_alignment', 'germline_alignment_aa', 'junction', 'junction_aa', 'np1', 'np1_aa', 'np2', 'np2_aa', 'np3', 'np3_aa', 'cdr1', 'cdr1_aa', 'cdr2', 'cdr2_aa', 'cdr3', 'cdr3_aa', 'fwr1', 'fwr1_aa', 'fwr2', 'fwr2_aa', 'fwr3', 'fwr3_aa', 'fwr4', 'fwr4_aa', 'v_score', 'v_identity', 'v_support', 'v_cigar', 'd_score', 'd_identity', 'd_support', 'd_cigar', 'd2_score', 'd2_identity', 'd2_support', 'd2_cigar', 'j_score', 'j_identity', 'j_support', 'j_cigar', 'c_score', 'c_identity', 'c_support', 'c_cigar', 'v_sequence_start', 'v_sequence_end', 'v_germline_start', 'v_germline_end', 'v_alignment_start', 'v_alignment_end', 'd_sequence_start', 'd_sequence_end', 'd_germline_start', 'd_germline_end', 'd_alignment_start', 'd_alignment_end', 'd2_sequence_start', 'd2_sequence_end', 'd2_germline_start', 'd2_germline_end', 'd2_alignment_start', 'd2_alignment_end', 'j_sequence_start', 'j_sequence_end', 'j_germline_start', 'j_germline_end', 'j_alignment_start', 'j_alignment_end', 'c_sequence_start', 'c_sequence_end', 'c_germline_start', 'c_germline_end', 'c_alignment_start', 'c_alignment_end', 'cdr1_start', 'cdr1_end', 'cdr2_start', 'cdr2_end', 'cdr3_start', 'cdr3_end', 'fwr1_start', 'fwr1_end', 'fwr2_start', 'fwr2_end', 'fwr3_start', 'fwr3_end', 'fwr4_start', 'fwr4_end', 'v_sequence_alignment', 'v_sequence_alignment_aa', 'd_sequence_alignment', 'd_sequence_alignment_aa', 'd2_sequence_alignment', 'd2_sequence_alignment_aa', 'j_sequence_alignment', 'j_sequence_alignment_aa', 'c_sequence_alignment', 'c_sequence_alignment_aa', 'v_germline_alignment', 'v_germline_alignment_aa', 'd_germline_alignment', 'd_germline_alignment_aa', 'd2_germline_alignment', 'd2_germline_alignment_aa', 'j_germline_alignment', 'j_germline_alignment_aa', 'c_germline_alignment', 'c_germline_alignment_aa', 'junction_length', 'junction_aa_length', 'np1_length', 'np2_length', 'np3_length', 'n1_length', 'n2_length', 'n3_length', 'p3v_length', 'p5d_length', 'p3d_length', 'p5d2_length', 'p3d2_length', 'p5j_length', 'v_frameshift', 'j_frameshift', 'd_frame', 'd2_frame', 'consensus_count', 'duplicate_count', 'umi_count', 'cell_id', 'clone_id', 'repertoire_id', 'sample_processing_id', 'data_processing_id', 'rearrangement_id', 'rearrangement_set_id', 'germline_database']

#turn tcr adata obs df into df with AIRR tsv columns
clonotype_df = tcr_adat.obs
airr_df = clonotype_df.melt(
    id_vars=["individual", "number_of_cells", "tcr", "TRAV", "TRAJ", "TRBV", "TRBJ"],
    value_vars=["CDR3a", "CDR3b"],
    var_name="chain_tmp",
    value_name="cdr3_aa"
)

# Map chain_tmp to locus (one locus type per row)
airr_df["locus"] = airr_df["chain_tmp"].map({
    "CDR3a": "TRA",
    "CDR3b": "TRB"
})

# Assign V/J
airr_df["v_call"] = np.where(airr_df["locus"]=="TRA", airr_df["TRAV"], airr_df["TRBV"])
airr_df["j_call"] = np.where(airr_df["locus"]=="TRA", airr_df["TRAJ"], airr_df["TRBJ"])

# use tcr as sequence id adding locus type to end
airr_df["sequence_id"] = airr_df["tcr"] + "_" + airr_df["locus"]

airr_df.rename(columns={
    "number_of_cells": "duplicate_count",
    "individual": "repertoire_id"
}, inplace=True)

airr_df.drop(columns=["TRAV", "TRAJ", "TRBV", "TRBJ", "chain_tmp","tcr"], inplace=True)

# Add any missing columns in airr_df with NaN
airr_df = airr_df.reindex(columns=col_names_airr)
airr_df
```

```python
#turn df individual IDs into str to avoid issues
df["Individual ID"] = df["Individual ID"].astype(str)

#get the individuals (both healthy and unhealthy) from studies that study the given target disease
def get_disease_individuals(target_disease: str, meta: pd.DataFrame):
    mask = (
        df
        .groupby("Author")["Disease Metatype"]
        .transform(lambda x: x.eq(target_disease).any())
    )

    result = df[mask]
    individual_list = (result.groupby(["Author", "Disease Metatype"])["Individual ID"].unique())
    print(individual_list)

    return individual_list.to_dict()
```

```python
#create AIRR tsv files per individual

def create_tsv_individual(disease_ind_dict, target_disease, airr_dataframe, meta_df, create_tsv=False, tsv_path=None):
    metadata_for_csv = pd.DataFrame(columns=["subject_id", "disease", "disease_metatype", "author", "disease_status","filename"])
   
    for (author, disease), individuals in disease_ind_dict.items():
        label = "positive" if disease == target_disease else "negative"
        
        for individual in individuals:
            individual = str(individual)

            #get the first list item because one individual may have several samples
            metadata_values = meta_df.loc[meta_df["Individual ID"]==individual, ["Individual ID", "Disease", "Disease Metatype", "Author"]].values.tolist()[0]
            
            #get the section relevant for the individual
            ind_df = airr_dataframe[airr_dataframe["repertoire_id"] == individual]
            file_name = f"{target_disease}_{label}_{individual}.tsv"

            #metadata for immuneml (this will be cocatenated with other disease metadata)
            metadata_values.extend([label,file_name])
            metadata_for_csv.loc[len(metadata_for_csv)] = metadata_values
            
            if create_tsv:
                ind_df.to_csv(tsv_path + file_name,sep="\t",index=False)

    return metadata_for_csv
```

```python
covid19_dict = get_disease_individuals("COVID-19", df)
```

```python
#use create_tsv=True once (create_tsv=False to just get the meta info)
covid19_meta = create_tsv_individual(
    disease_ind_dict=covid19_dict,
    target_disease="COVID-19",
    airr_dataframe=airr_df,
    meta_df=df,
    create_tsv=False,
    tsv_path="/vol/data/immuneML/tcr_atlas_data/covid19_leukemia/repertoires/")

covid19_meta
```
